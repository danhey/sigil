```python
import jax

import jax.numpy as jnp
from diffrax import diffeqsolve, ODETerm, Kvaerno3
from diffrax import diffeqsolve, Dopri5, ODETerm, SaveAt, PIDController, Tsit5

jax.config.update("jax_enable_x64", True)
```

```python
from sigil.constants import Msun, Lsun#, Tsun

M = 1
m = 0.01 * Msun
mf = 0.8 * M # interaction point for shooting
X = 0.7
Y = 0.28
```

```python
from sigil.shootf import init_guess, load1, load2
from sigil.equations import drdm, dldm, dPdm, dTdm
from sigil.physics import mu_from_composition, rho_ideal
from sigil.opacity import opacity

R, L, Pc, Tc = init_guess(M, X, Y)
y_center = load1(m, Pc, Tc, X, Y)
y_surface = load2(M, R, Lsun, X, Y)
```

```python
from numpy import genfromtxt
from sigil.opacity import opacity

op = genfromtxt('data/opacity/solar.txt', invalid_raise=False)

logT = jnp.asarray(op[:,0])
logR = jnp.arange(-8, 1.5, step=0.5)
logkappa = jnp.asarray(op[:, 1:])

opacity(1,1,logT, logR, logkappa)
```

```python
def deriv(m, u, args):
    r, l, P, T = u
    X, Y = args
    
    mu = mu_from_composition(X, Y)
    rho = rho_ideal(P, T, mu)
    # kappa = opacity()
    kappa = opacity(rho, T, logT, logR, logkappa)
    du = jnp.stack([
        drdm(r, rho),
        dldm(T, rho, X, Y),
        dPdm(m, r),
        dTdm(m, r, l, P, T, kappa)
    ])
    return du
```

```python
def lorenz(t, u, args):
    return jnp.stack([10.0*(u[1]-u[0]), u[1]*(28.0-u[2]) - u[1], u[0]*u[1] - (8/3)*u[2]])
```
